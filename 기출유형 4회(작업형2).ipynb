## Q. [마케팅] 자동차 시장 세분화
- 자동차 회사는 새로운 전략을 수립하기 위해 4개의 시장으로 세분화했습니다.
- 기존 고객 분류 자료를 바탕으로 신규 고객이 어떤 분류에 속할지 예측해주세요!


- 예측할 값(y): "Segmentation" (1,2,3,4)
- 평가: Macro f1-score
- data: train.csv, test.csv
- 제출 형식:
~~~
ID,Segmentation
458989,1
458994,2
459000,3
459003,4
~~~

### 답안 제출 참고
- 아래 코드 예측변수와 수험번호를 개인별로 변경하여 활용
- pd.DataFrame({'ID': test.ID, 'Segmentation': pred}).to_csv('003000000.csv', index=False)

### 노트북 구분
- basic: 수치형 데이터만 활용 -> 학습 및 test데이터 예측
- intermediate: 범주형 데이터도 활용 -> 학습 및 test데이터 예측
- advanced: 학습 및 교차 검증(모델 평가) -> 하이퍼파라미터 튜닝 -> test데이터 예측

```python
# 내 코드

import pandas as pd

train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

# #탐색
# pd.get_option('display.max_column')

# # print(train.head())
# # print(test.head())
# import sklearn.preprocessing
# print((sklearn.preprocessing.MinMaxScaler))


target =train.pop('Segmentation')

train=train.drop('ID',axis =1)

test_ID = test.pop('ID')

train.columns
train.info()
from sklearn.preprocessing import MinMaxScaler
scaler =MinMaxScaler()
n_cols = ['Age', 'Work_Experience', 'Family_Size']
train[n_cols] = scaler.fit_transform(train[n_cols])
test[n_cols] = scaler.transform(test[n_cols])
# print(train.head())

from sklearn.preprocessing import LabelEncoder # 하나씩 해야 함

print(train.shape)
c_cols=['Gender', 'Ever_Married', 'Graduated', 'Profession','Spending_Score', 'Var_1']
train=pd.get_dummies(train[c_cols])
test=pd.get_dummies(test[c_cols])
print(train.shape)


from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val =  train_test_split(train,target,test_size=0.2,random_state=0)




#모델
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0)
rf.fit(X_tr,y_tr)
pred=rf.predict(X_val)#  y 값은 pred 에 저장되는거니깐 빼야함 하나만 넣기

#만든 모델 평가 지표
from sklearn.metrics import accuracy_score
accuracy_score(y_val,pred)

from sklearn.metrics import f1_score
print(f1_score(y_val,pred,average='macro'))


# 실제 test 모델 돌리고 예측값 저장

pred=rf.predict(test)

pd.DataFrame({'ID': test_ID, 'Segmentation': pred}).to_csv('003000000.csv', index=False)

pd.read_csv('003000000.csv')
```
