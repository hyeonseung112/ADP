- intermediate: Î≤îÏ£ºÌòï Îç∞Ïù¥ÌÑ∞ÎèÑ ÌôúÏö© -> ÌïôÏäµ Î∞è testÎç∞Ïù¥ÌÑ∞ ÏòàÏ∏°

```python
# ÎÇ¥ ÏΩîÎìú

import pandas as pd

train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

# #ÌÉêÏÉâ
# pd.get_option('display.max_column')

# # print(train.head())
# # print(test.head())
# import sklearn.preprocessing
# print((sklearn.preprocessing.MinMaxScaler))


target =train.pop('Segmentation')

train=train.drop('ID',axis =1)

test_ID = test.pop('ID')

train.columns
train.info()
from sklearn.preprocessing import MinMaxScaler
scaler =MinMaxScaler()
n_cols = ['Age', 'Work_Experience', 'Family_Size']
train[n_cols] = scaler.fit_transform(train[n_cols])
test[n_cols] = scaler.transform(test[n_cols])
# print(train.head())

from sklearn.preprocessing import LabelEncoder # ÌïòÎÇòÏî© Ìï¥Ïïº Ìï®

print(train.shape)
c_cols=['Gender', 'Ever_Married', 'Graduated', 'Profession','Spending_Score', 'Var_1']
train=pd.get_dummies(train[c_cols])
test=pd.get_dummies(test[c_cols])
print(train.shape)


from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val =  train_test_split(train,target,test_size=0.2,random_state=0)




#Î™®Îç∏
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0)
rf.fit(X_tr,y_tr)
pred=rf.predict(X_val)#  y Í∞íÏùÄ pred Ïóê Ï†ÄÏû•ÎêòÎäîÍ±∞ÎãàÍπê ÎπºÏïºÌï® ÌïòÎÇòÎßå ÎÑ£Í∏∞

#ÎßåÎì† Î™®Îç∏ ÌèâÍ∞Ä ÏßÄÌëú
from sklearn.metrics import accuracy_score
accuracy_score(y_val,pred)

from sklearn.metrics import f1_score
print(f1_score(y_val,pred,average='macro'))


# Ïã§Ï†ú test Î™®Îç∏ ÎèåÎ¶¨Í≥† ÏòàÏ∏°Í∞í Ï†ÄÏû•

pred=rf.predict(test)

pd.DataFrame({'ID': test_ID, 'Segmentation': pred}).to_csv('003000000.csv', index=False)

pd.read_csv('003000000.csv')
```
